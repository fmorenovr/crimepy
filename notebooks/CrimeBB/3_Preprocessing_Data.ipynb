### Summarizing CrimeBB

```python
website_df = pd.read_csv(f"{CSV_PROCESSED}sites.csv", sep="\t", low_memory=False)
boards_df = pd.read_csv(f"{CSV_PROCESSED}boards.csv", sep="\t", low_memory=False)
threads_df = pd.read_csv(f"{CSV_PROCESSED}threads.csv", sep="\t", low_memory=False)
```

```python
chunk_size = 1000000

posts_reader = pd.read_csv(f"{CSV_PROCESSED}posts.csv", sep="\t", low_memory=False, iterator=True)
            
posts_df = pd.DataFrame()

len_readed=chunk_size
while len_readed>=chunk_size:
    cur_posts_df = posts_reader.get_chunk(chunk_size).copy()
    cur_posts_df.drop_duplicates(inplace=True)

    posts_df = pd.concat([posts_df, cur_posts_df], ignore_index=True)

    len_readed = cur_posts_df.shape[0]
```

```python
posts_df
```

#### Merging

```python
posts_website_df = pd.merge(posts_df, website_df, how="left", on="site_id")
```

```python
posts_boards_df = pd.merge(posts_website_df, boards_df[["site_id", "board_id", "board_title"]].drop_duplicates(), on=["site_id", "board_id"], how="left")
```

```python
posts_threads_df = pd.merge(posts_boards_df, threads_df[["site_id", "board_id", "thread_id", "thread_title"]].drop_duplicates(), on=["site_id", "board_id", "thread_id"], how="left")
```

```python
crimebb_df = posts_threads_df[['post_id', 'site_id', 'board_id', 'thread_id', 'user_id', 
                         'site_name', 'board_title', 'thread_title', 'username', 'content', 
                         'user_reputation', 'post_data_creation']].copy()
crimebb_df
```

```python
crimebb_df.to_csv(f"{CSV_PROCESSED}crimeBB_{YEAR}.csv", sep='\t', index=False)
```

### Drawing Missing Values

#### user_id equals None

```python
none_members_users = members_df[ (members_df["username"].str.lower()=="none") | (members_df["username"].isna()) ].copy().drop_duplicates()
```

```python
none_threads_users = threads_df[ (threads_df["username"].str.lower()=="none") | (threads_df["username"].isna()) ].copy().drop_duplicates()
```

```python
none_posts_users = posts_df[ (posts_df["username"].str.lower()=="none") | (posts_df["username"].isna()) ].copy().drop_duplicates()
```

#### thread title None

```python
threads_no_none_users_df = threads_df[ (threads_df["username"].str.lower()!="none") & (~threads_df["username"].isna()) ].copy().drop_duplicates()
```

```python
none_threads_title = threads_no_none_users_df[threads_no_none_users_df["thread_title"].isna()]
```

#### posts content None

```python
posts_no_none_users_df = posts_df[ (posts_df["username"].str.lower()!="none") & (~posts_df["username"].isna()) ].copy().drop_duplicates()
```

```python
none_posts_content = posts_no_none_users_df[posts_no_none_users_df["content"].isna()]
```

#### Draw

```python
import matplotlib.pyplot as plt

legends_ = ["members", "threads", "posts"]
values_ = [len(none_members_users), len(none_threads_users), len(none_posts_users)]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(legends_,values_, label='Null Usernames')
for x,y,p in zip(legends_, values_, values_):
  plt.text(x, y, p)

ax.bar(["posts "], [len(none_threads_title)], label='Null Thread_title')
for x,y,p in zip(["posts "], [len(none_threads_title)], [len(none_threads_title)]):
  plt.text(x, y, p)

ax.bar(["posts  "], [len(none_posts_content)], label='Null Post_content')
for x,y,p in zip(["posts  "], [len(none_posts_content)], [len(none_posts_content)]):
  plt.text(x, y, p)

ax.legend()
ax.set_title("Missing values per table")
ax.set_ylabel("Number of missing values")
plt.show()
```
