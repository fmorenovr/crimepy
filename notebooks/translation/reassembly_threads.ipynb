```python
import pandas as pd
```

```python
posts = pd.read_csv("posts.csv")
translated = pd.read_csv("translated_posts.csv")

posts = posts.set_index('IdPost')
translated = translated.set_index('IdPost')
```

```python
print("Original size: %d" % posts.shape[0])
print("Translated size: %d" % translated.shape[0])
```

```python
posts = posts.sort_values(by=['IdThread', 'IdPost'])
posts.head()
```

```python
posts['Content_en'] = posts.index.map(lambda idPost: translated.loc[idPost, 'Content'] if idPost in translated.index else 0)
```

```python
posts.to_csv("posts_translated.csv")
```

```python
posts['Content_en'] == posts['Content_en'].astype(str)
```
